```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
%matplotlib inline
```

```python
# Find Mean, Median, and Mode
def summary_stats(data):
    print('Average: \t{}'.format(np.mean(data)))
    print('Median: \t{}'.format(np.median(data)))
    
    print('Standard Dev: \t{}'.format(np.std(data)))    
    print('Highbound: \t{}'.format(np.mean(data) + np.std(data)))
    print('Lowbound: \t{}'.format(np.mean(data) - np.std(data))) 
    try:
        "Mode: \t{}".format(statistics.mode(data))
    except:
        "Mode: \t{}".format('Nan')
    
    print('Variance: \t{}'.format(np.var(data)))
    print('Standard Error:\t{}'.format(np.std(data)/len(data)))
```

```python
# Generate beta distribution with p =0.5 beta(shape,scale,size)
def distribution_summary(distribution_data):
    data=distribution_data.iloc[:,0]
    name=distribution_data.columns[0]
    fig = plt.figure(figsize=(4,3))
    plt.hist(data)
    plt.title(name)
    
    
    # Add a vertical line at the mean
    plt.axvline(data.mean(), color='c', linestyle='solid', linewidth=2, label=data.mean)

    # Add a vertical line at one standard deviatin above the mean
    plt.axvline(data.mean() + data.std(), color='c', linestyle='dashed', linewidth=2)

    # Add a verticalline at on standard deviation below the mean
    plt.axvline(data.mean()-data.std(), color='c', linestyle='dashed', linewidth=2)
    plt.show()
    # Print the histogram
    #summary = summary_stats(data)
    #output = [fig, summary]
    #return output
    summary_stats(data)
```

```python
# Generate beta distribution with p =0.5 beta(shape,scale,size)
beta = pd.DataFrame({'beta': np.random.beta(2,.5,100)})
distribution_summary(beta)

# Generate logseries distribution
logseries = pd.DataFrame({'logseries': np.random.logseries(.8,100)})
distribution_summary(logseries)

# Generate beta distribution
logistic = pd.DataFrame({'logistic': np.random.logistic(8,.8,100)})
distribution_summary(logistic)

# Generate pareto distribution 
pareto = pd.DataFrame({'pareto':np.random.pareto(6,100)})
distribution_summary(pareto)

# Generate multinomial distribution
gumbel = pd.DataFrame({'gumbel': np.random.gumbel(0, 3, 100)})
distribution_summary(gumbel)

# Generate Uniform Distribution
uniform = pd.DataFrame({'uniform': np.random.uniform(low=0.0, high=1.0, size=100)})
distribution_summary(uniform)
```

```python
# 1. Generate 2 normally distributed variablesm, mean = 5, st dev =.5 & mean = 10, st dev=1, 
# np.random.normal(mean, sd, count)
norm1 = np.random.normal(5,.5,100)
norm2 = np.random.normal(10,1,100)

# 2. Add the 2 variables together to create a 3rd variable
norm3 = norm1 + norm2
```

```python
# 3. Graph the 3rd variable using a histogram
plt.hist(norm3)
plt.title('norm3')

# 4. computer the mean and standard deviation
mean = np.mean(norm3)
st_dev = np.std(norm3)
lowbound = mean - st_dev
highbound = mean + st_dev

# Add a vertical line at the mean
plt.axvline(mean, color='c', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviatin above the mean
plt.axvline(lowbound, color='c', linestyle='dashed', linewidth=2)

# Add a verticalline at on standard deviation below the mean
plt.axvline(highbound, color='c', linestyle='dashed', linewidth=2)

plt.show()
summary_stats(norm3)

print("""
Results vary depending on seed but generally adding two datasets with
different probabilities introduced lopsided variance""")
```
